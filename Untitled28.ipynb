```python
df=pd.read_csv("Loan_Data.csv");
```

```python
print(df.info())
```

```python
df["Dependents"] = df["Dependents"].fillna(df["Dependents"].mode()[0])
df["Education"]=df["Education"].fillna(df["Education"].mode()[0])
df["Self_Employed"] = df["Self_Employed"].fillna(df["Self_Employed"].mode()[0])
df["LoanAmount"]=df["LoanAmount"].fillna(df["LoanAmount"].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
```

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["Married"]=df["Married"].fillna(df["Married"].mode()[0])
df["Credit_History"]=df["Credit_History"].fillna(df["Credit_History"].mode()[0])
```

```python
print(df.isnull().sum())
```

```python
df=df.drop_duplicates()
df=df.drop(columns=["Loan_ID"]);
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
catcol=df.select_dtypes(include="object").columns;
print(catcol);
```

```python
for i in catcol:
  df[i]=le.fit_transform(df[i]);
print(df.head(10))
```

```python
x=df.drop(columns="Loan_Status");
y=df['Loan_Status'];
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.2,random_state=42);
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mode=RandomForestClassifier(max_depth=3,n_estimators=100,random_state=42)
```

```python
mode.fit(xtr,ytr);
pred=mode.predict(xts);
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score;
print("accuracy=",accuracy_score(yts,pred))
```

```python
print("mse",mean_squared_error(yts,pred))
print("mae",mean_absolute_error(yts,pred))
```
